```python
import pandas as pd
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn import preprocessing
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
```

```python
pigfoot1 = pd.read_excel('data_pigfoot/data_1.xls',header=None)
pigfoot2 = pd.read_excel('data_pigfoot/data_2.xls',header=None)
pigfoot3 = pd.read_excel('data_pigfoot/data_3.xls',header=None)
pigfoot4 = pd.read_excel('data_pigfoot/data_4.xls',header=None)
pigfoot5 = pd.read_excel('data_pigfoot/data_5.xls',header=None)
pigfoot6 = pd.read_excel('data_pigfoot/data_6.xls',header=None)
pigfoot7 = pd.read_excel('data_pigfoot/data_7.xls',header=None)
pigfoot8 = pd.read_excel('data_pigfoot/data_8.xls',header=None)
pigfoot9 = pd.read_excel('data_pigfoot/data_9.xls',header=None)
pigfoot10 = pd.read_excel('data_pigfoot/data_10.xls',header=None)
pigfoot11 = pd.read_excel('data_pigfoot/data_11.xls',header=None)
pigfoot12 = pd.read_excel('data_pigfoot/data_12.xls',header=None)
```

```python
frames = [pigfoot1,pigfoot2,pigfoot3,pigfoot4,pigfoot5,pigfoot6,pigfoot7,pigfoot8,pigfoot9,pigfoot10,pigfoot11,pigfoot12]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
# 각 column에 indexing
result.columns = ["total_number", "house", "call", "morning", "lunch", "dinner","yasik"]
print(result.head())
print(result)
```

```python
# 한달에 한 가게가 받는 콜 수 column 추가
result['call_for_onestore'] = result['call']/result['total_number']
```

```python
# Normalization으로 Min-Max scaler 사용
min_max_scaler = preprocessing.MinMaxScaler()
x = result['total_number'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['total_number'] = pd.DataFrame(x_scaled)
```

```python
x = result['house'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['house'] = pd.DataFrame(x_scaled)
```

```python
# Multivariable regression
LP_result = sm.ols(formula = 'call_for_onestore ~ total_number + house + morning + lunch + dinner + yasik', data = result).fit()
```

```python
print(LP_result.summary())
```
