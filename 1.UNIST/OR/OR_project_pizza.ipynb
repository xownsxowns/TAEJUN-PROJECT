```python
import pandas as pd
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn import preprocessing
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
```

```python
pizza1 = pd.read_excel('data_pizza/data_1.xls',header=None)
pizza2 = pd.read_excel('data_pizza/data_2.xls',header=None)
pizza3 = pd.read_excel('data_pizza/data_3.xls',header=None)
pizza4 = pd.read_excel('data_pizza/data_4.xls',header=None)
pizza5 = pd.read_excel('data_pizza/data_5.xls',header=None)
pizza6 = pd.read_excel('data_pizza/data_6.xls',header=None)
pizza7 = pd.read_excel('data_pizza/data_7.xls',header=None)
pizza8 = pd.read_excel('data_pizza/data_8.xls',header=None)
pizza9 = pd.read_excel('data_pizza/data_9.xls',header=None)
pizza10 = pd.read_excel('data_pizza/data_10.xls',header=None)
pizza11 = pd.read_excel('data_pizza/data_11.xls',header=None)
pizza12 = pd.read_excel('data_pizza/data_12.xls',header=None)
```

```python
frames = [pizza1, pizza2, pizza3, pizza4, pizza5, pizza6, pizza7, pizza8, pizza9, pizza10, pizza11, pizza12]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
# 각 column에 indexing
result.columns = ["total_number", "house", "call", "morning", "lunch", "dinner","yasik"]
print(result.head())
print(result)
```

```python
# 한달에 한 가게가 받는 콜 수 column 추가
result['call_for_onestore'] = result['call']/result['total_number']
```

```python
# Normalization으로 Min-Max scaler 사용
min_max_scaler = preprocessing.MinMaxScaler()
x = result['total_number'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['total_number'] = pd.DataFrame(x_scaled)
```

```python
x = result['house'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['house'] = pd.DataFrame(x_scaled)
```

```python
# Multivariable regression
LP_result = sm.ols(formula = 'call_for_onestore ~ total_number + house + morning + lunch + dinner + yasik', data = result).fit()
```

```python
print(LP_result.summary())
```
