```python
import pandas as pd
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn import preprocessing
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
chicken1 = pd.read_excel('data_chicken/data_1.xls',header=None)
chicken2 = pd.read_excel('data_chicken/data_2.xls',header=None)
chicken3 = pd.read_excel('data_chicken/data_3.xls',header=None)
chicken4 = pd.read_excel('data_chicken/data_4.xls',header=None)
chicken5 = pd.read_excel('data_chicken/data_5.xls',header=None)
chicken6 = pd.read_excel('data_chicken/data_6.xls',header=None)
chicken7 = pd.read_excel('data_chicken/data_7.xls',header=None)
chicken8 = pd.read_excel('data_chicken/data_8.xls',header=None)
chicken9 = pd.read_excel('data_chicken/data_9.xls',header=None)
chicken10 = pd.read_excel('data_chicken/data_10.xls',header=None)
chicken11 = pd.read_excel('data_chicken/data_11.xls',header=None)
chicken12 = pd.read_excel('data_chicken/data_12.xls',header=None)
```

```python
frames = [chicken1,chicken2,chicken3,chicken4,chicken5,chicken6,chicken7,chicken8,chicken9,chicken10,chicken11,chicken12]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
# 각 column에 naming
result.columns = ["total_number", "house", "call", "morning", "lunch", "dinner","yasik"]
print(result.head())
print(result)
```

```python
result
```

```python
# 한달에 한 가게가 받는 콜 수 column 추가
result['call_for_onestore'] = result['call']/result['total_number']
```

```python
result
```

```python
# Normalization으로 Min-Max scaler 사용
min_max_scaler = preprocessing.MinMaxScaler()
x = result['total_number'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['total_number'] = pd.DataFrame(x_scaled)
```

```python
x = result['house'].values.astype(float)
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
result['house'] = pd.DataFrame(x_scaled)
```

```python
result
```

```python
# Multivariable regression
LP_result = sm.ols(formula = 'call_for_onestore ~ total_number + house + morning + lunch + dinner + yasik', data = result).fit()
```

```python
print(LP_result.summary())
```
