```python
import mne
from mne.preprocessing import ICA, create_ecg_epochs
from mne.datasets import sample
```

```python
# 1. low-pass the data with mne.io.Raw.filter()
# 2. extract epochs with mne.Epochs
# 3. Decimate the Epochs object using mne.Epochs.decimate() or the decim argument to the mne.Epochs object.
data = mne.io.read_raw_edf('sub1_1_filtered.edf', preload=True)
```

```python
visual = mne.viz.plot_raw(data)
```

```python
data.filter(1, 30, fir_design='firwin')
```

```python
import numpy as np
data.plot_psd(tmax=np.inf, fmax=250)
```

```python
# Removing power-line noise with notch filtering
data.notch_filter(np.arange(60,150,60), filter_length='auto', phase='zero')
```

```python
data.plot_psd(area_mode='range', tmax=10,average=False)
```

```python
# Removing power-line noise with low-pass filtering
# Low pass filtering below 50 Hz
data.filter(None, 50., fir_design = 'firwin')
data.plot_psd(area_mode='range', tmax=10,average=False)
```

```python
# High-pass filtering to remove slow drifts
data.filter(1., None, fir_design = 'firwin')
data.plot_psd(area_mode='range', tmax=10.0, average=False)
```

```python
# Downsampling and decimation
data.resample(100, npad='auto') # set sampling frequency to 100Hz
data.plot_psd(area_mode='range', tmax=10.0)
```
