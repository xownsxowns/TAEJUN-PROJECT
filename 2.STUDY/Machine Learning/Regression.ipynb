```python
# 회귀는 연속적인 값을 예측하는데 쓰임
import sklearn.datasets
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import normalize
from sklearn.metrics import r2_score
```

```python
diabetes = sklearn.datasets.load_diabetes()
X, Y = normalize(diabetes['data']), diabetes['target']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.8)
linear = LinearRegression()
linear.fit(X_train,Y_train)
preds_linear = linear.predict(X_test)
corr_linear = round(pd.Series(preds_linear).corr(pd.Series(Y_test)),3)
rsquared_linear = r2_score(Y_test, preds_linear)
```

```python
print("선형 계수: ")
print(linear.coef_)
plt.scatter(preds_linear, Y_test)
plt.title("Regression result. corr = {0} R^2 score = {1}".format(corr_linear, rsquared_linear))
plt.xlabel("Prediction value")
plt.ylabel("Real value")
plt.plot(Y_test, Y_test, 'k--')
plt.show()
```

```python
lasso = Lasso()
lasso.fit(X_train, Y_train)
preds_lasso = lasso.predict(X_test)
corr_lasso = round(pd.Series(preds_lasso).corr(pd.Series(Y_test)),3)
rsquared_lasso = round(r2_score(Y_test, preds_lasso), 3)
print("라소 계수: ")
print(lasso.coef_)
plt.scatter(preds_lasso, Y_test)
plt.title("lasso regression result. corr={0} R^2 score={1}".format(corr_lasso, rsquared_lasso))
plt.xlabel("Prediction value")
plt.ylabel("Real value")
plt.plot(Y_test, Y_test, 'k--')
plt.show()
```

```python
# 최소 제곱법은 loss function을 최소화 하는 계수를 찾는 것
# 비선형 커브피팅
# 아래 예제는 y = 2+3x^2 함수에 약간의 잡음을 섞은 데이터를 만들고 curve_fit 함수로 결과를 검증한것
from scipy.optimize import curve_fit
import numpy as np
xs = np.array([1.0, 2.0, 3.0, 4.0])
ys = 2.0 + 3.0 * xs * xs + 0.2 * np.random.uniform(3)
def calc(x, a, b):
    return a+b*x*x
cf = curve_fit(calc, xs, ys)
best_fit_params = cf[0]
print(best_fit_params)
```
