```python
# 간단한 선그래프
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
# x축에 연도, y축에 GDP가 있는 선 그래프를 만들자
plt.plot(years, gdp, color='green',marker='o',linestyle='solid')
```

```python
# 제목을 더하자
plt.title("Nominal GDP")
# y축에 레이블을 추가하자
plt.ylabel("Billions of $")
plt.show()
```

```python
# 막대 그래프
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
# 막대 너비의 기본값이 0.8 이므로 막대가 가운데로 올 수 있도록 왼쪽 좌표에 0.1씩 더해 주자
# enumerate > 열거하다
xs = [i + 0.1 for i, _ in enumerate(movies)]
```

```python
# 왼편으로부터 x축의 위치가 xs이고 높이가 num_oscars인 막대를 그리자
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
```

```python
# 아이리스 데이터셋 사용
# 각 종마다 50개씩 총 150개의 데이터 표본이 있고 측정값은 붓꽃의 꽃잎과 꽃받침의 길이 및 너비
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets

def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'], columns = ds['feature_names'])
    code_species_map = dict(zip(range(3), ds['target_names']))
    df['species'] = [code_species_map[c] for c in ds['target']]

df = get_iris_df()
```

```python
ds = sklearn.datasets.load_iris()
df = pd.DataFrame(ds['data'],columns = ds['feature_names'])
code_species_map = dict(zip(range(3),ds['target_names']))
df['species'] = [code_species_map[c] for c in ds['target']]
```

```python
# 원형차트
sums_by_species = df.groupby('species').sum()
print(sums_by_species)
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='pie', fontsize=10)
plt.ylabel(var, horizontalalignment='left')
plt.title(var + ' classifying Iris', fontsize=25)
plt.show()
```

```python
# 원형차트 subplot
sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind = 'pie', subplots = True, layout=(2,2), legend = False)
plt.title('Overall measured value following by species')
plt.show()
```

```python
# 막대그래프
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind = 'bar', fontsize=15, rot=30)

plt.title(var+' classifying Iris', fontsize = 20)
plt.show()

sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind = 'bar', subplots = True, fontsize=12)

plt.title('Overall measured value following by species')
plt.show()
```

```python
# 히스토그램
df.plot(kind='hist', subplots=True, layout=(2,2))
plt.suptitle('Iris histogram', fontsize=20)
plt.show()
```

```python
# unique는 유일값 찾기
for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(kind = 'hist', alpha=0.4, label=spec)

plt.legend(loc='upper right')
plt.suptitle('petal length by Iris')
plt.show()
```

```python
# fundamental statistical analysis
# 대푯값에는 평균&표준편차, 중간값&25%백분위수&75%백분위수
col = df['petal length (cm)']
average = col.mean()
std = col.std()
median = col.quantile(0.5)
percentile25 = col.quantile(0.25)
percentile75 = col.quantile(0.75)
```

```python
# 이상치가 있으면 데이터 분포 자체가 이상해지니 아래 코드처럼 하위 25% 및 상위 25%를 걸러내고 나머지 데이터로만 평균을 구하는 방식으로 상요하기도함
col = df['petal length (cm)']
perc25 = col.quantile(0.25)
perc75 = col.quantile(0.75)
clean_avg = col[(col > perc25) & (col < perc75)].mean()
```

```python
# Box plot
col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i: i%50)
df.pivot('ind','species')[col].plot(kind='box')
plt.show()
```

```python
# scatter
# c:산포도 점의 색깔, s:산포도 점의 크기, marker:점의 모양, alpha:투명도
df.plot(kind="scatter", x="sepal length (cm)", y="sepal width (cm)")
plt.title("length vs. width")
plt.show()
```

```python
plt.close()
colors = ["r","g","b"]
markers = [".","*","^"]
fig, ax = plt.subplots(1,1)
for i, spec in enumerate(df['species'].unique()):
    ddf = df[df['species']==spec]
    ddf.plot(kind="scatter",
            x = "sepal width (cm)", y = "sepal length (cm)",
            alpha = 0.5, s=10*(i+1), ax=ax,
            color = colors[i],marker=markers[i], label=spec)

plt.legend()
plt.show()
```

```python
# 산포도와 로그 축
import pandas as pd
import sklearn.datasets as ds
import matplotlib.pyplot as plt

# 팬더스 데이터프레임 생성
bs = ds.load_boston()
df = pd.DataFrame(bs.data, columns=bs.feature_names)
df['MEDV'] = bs.target

# 일반적인 산포도
df.plot(x='CRIM', y='LSTAT', kind='scatter')
plt.title('Crime rate')
plt.show()
```

```python
# 로그축
df.plot(x='CRIM', y='LSTAT', kind = 'scatter', logx=True)
plt.title('Crime rate(log)')
plt.show()
```

```python
# 산포 행렬
plt.close()
from pandas.plotting import scatter_matrix
ds = sklearn.datasets.load_iris()
df = pd.DataFrame(ds['data'],columns = ds['feature_names'])
code_species_map = dict(zip(range(3),ds['target_names']))
df['species'] = [code_species_map[c] for c in ds['target']]
scatter_matrix(df)
plt.show()
```

```python
# 히트맵
plt.close()
df.plot(kind="hexbin", x = "petal width (cm)", y="petal length (cm)")
plt.show()
```

```python
# 상관관계
# 피어슨 상관관계 & 순서형 상관관계
# 0.4이상=상관관계 고려, 0.7이상=어떤 변수의 변화를 다른 변수로 설명할 수 있는 확률이 50퍼 넘음, 0.95이상=거의 똑같다고 취급
df["sepal width (cm)"].corr(df["sepal length (cm)"]) #피어슨 상관관계
df["sepal width (cm)"].corr(df["sepal length (cm)"], method = "pearson") #피어슨 상관관계
df["sepal width (cm)"].corr(df["sepal length (cm)"], method = "spearman")
df["sepal width (cm)"].corr(df["sepal length (cm)"], method = "kendall")
```
