## StatsModels package

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
ar = np.r_[1, -0.75, 0.25]
ma = np.r_[1, 0.65, 0.35]
m = sm.tsa.ArmaProcess(ar, ma)
```

### generate_sample( ) : 주어진 ARMA 모형에서 나올 수 있는 시계열을 무작위로 생성

```python
np.random.seed(1234)
x = m.generate_sample(100)
plt.plot(x)
plt.show()
```

```python
plt.subplot(311)
plt.stem(m.acovf(10)) # 이론적인 자기 공분산(Autocovariance) 값
plt.subplot(312) 
plt.stem(m.acf(10)) # 이론적인 자기 상관계
plt.subplot(313)
plt.stem(m.pacf(10))
plt.show()
```
