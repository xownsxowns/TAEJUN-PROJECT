```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

## Filtering the Numerical values only

```python
str_list = []
for col_name, col_value in movie_summer.iteritems():
    if type(col_value[1]) == str:
        str_list.append(col_name)
```

```python
num_list = movie_summer.columns.difference(str_list)
```

```python
movie_num = movie_summer[num_list]
movie_num.head()
```

```python
movie_num = movie_num.fillna(value=0,axis=1)
movie_num.drop(movie_num.columns[0],axis=1,inplace=True)
movie_num.drop(movie_num.columns[0],axis=1,inplace=True)
```

```python
movie_num
```

```python
X = movie_num.values
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(X)
```

```python
x.shape
```

```python
ks = range(1,10)
inertias = []
for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(x)
    inertias.append(model.inertia_)

# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
# Set a 3 KMeans clustering
kmeans = KMeans(n_clusters=4)
# Compute cluster centers and predict cluster indices
X_clustered = kmeans.fit_predict(x)

# Define our own color map
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b', 3: 'y'}
label_color = [LABEL_COLOR_MAP[l] for l in X_clustered]

# Plot the scatter digram
plt.figure(figsize = (7,7))
plt.scatter(x[:,5],x[:,3], c= label_color, alpha=0.5)
plt.show()
```

```python
df = pd.DataFrame(x)
df = df[[0,1,2,3,4,5]] # only want to visualise relationships between first 3 projections
df['X_cluster'] = X_clustered
```

```python
df = df.rename(columns={0:'Seoul viewer',1:'Seoul sales',2:'movie price',3:'viewers', 4:'sales',5:'screens'})
df['quater'] = movie_summer['분기']
```
