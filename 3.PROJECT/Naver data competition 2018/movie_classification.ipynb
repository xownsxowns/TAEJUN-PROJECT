# 영화 데이터를 분석해보자

```python
import pandas as pd
```

```python
movie = pd.read_excel('movie_data.xlsx')
```

```python
movie
```

```python
movie['개봉일'] = pd.to_datetime(movie['개봉일'])
movie['month'] = pd.DatetimeIndex(movie['개봉일']).month
movie.head()
```

```python
cols = [0, 4]
movie.drop(movie.columns[cols],axis=1,inplace=True)
```

```python
movie.dropna(inplace=True)
```

```python
str_list = []
for col_name, col_value in movie.iteritems():
    if type(col_value[1]) == str:
        str_list.append(col_name)
```

```python
num_list = movie.columns.difference(str_list)
movie_num = movie[num_list]
movie_num.head()
```

```python
movie_num = movie_num.fillna(value=0,axis=1)
movie_num.drop(movie_num.columns[0],axis=1,inplace=True)
movie_num.drop(movie_num.columns[0],axis=1,inplace=True)
movie_num.head()
```

```python
X = movie_num.values
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(X)
```

```python
normalized_x = pd.DataFrame(x)
normalized_x = normalized_x.rename(columns={0:'Seoul viewer',1:'Seoul sales',2:'movie price',3:'viewers', 4:'sales',5:'screens'})
```

```python
normalized_x['quater'] = movie['분기']
normalized_x.head()
```

```python
normalized_x_onehot = normalized_x.copy()
normalized_x_onehot = pd.get_dummies(normalized_x, columns=['quater'], prefix = ['quater'])
print(normalized_x_onehot.head())
```

```python
normalized_x_onehot = pd.DataFrame(normalized_x_onehot)
normalized_x_onehot.head()
```

```python
from sklearn import linear_model
```

```python
x = normalized_x_onehot.drop(normalized_x_onehot.columns[3],axis=1)
```

```python
y = normalized_x_onehot['viewers']
y.head()
```

```python
from sklearn.model_selection import train_test_split,KFold,cross_val_score,cross_val_predict
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
lm = linear_model.LinearRegression()
```

```python
model = lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
confidence = lm.score(X_test,y_test)
```

```python
print("The linear cofficients",model.coef_)
```

```python
print("This is predicted values", predictions)
```

```python
print("This is the prediction accuracy", confidence)
```

### 랜덤포레스트

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

### Stochastic Gradient Descent Classifier

```python
sgd = SGDClassifier(penalty=None)
sgd.fit(X_train, y_train)
pred_sgd = sgd.predict(X_test)
```

```python
print(classification_report(y_test, pred_sgd))
```

### SVM

```python
svc = SVC()
svc.fit(X_train, y_train)
pred_svc = svc.predict(X_test)
```

```python
print(classification_report(y_test, pred_svc))
```

```python
#Finding best parameters for our SVC model
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=10)
```

```python
grid_svc.fit(X_train, y_train)
```

```python
#Best parameters for our svc model
grid_svc.best_params_
```

```python
#Let's run our SVC again with the best parameters.
svc2 = SVC(C = 1.2, gamma = 0.1, kernel= 'linear')
svc2.fit(X_train, y_train)
pred_svc2 = svc2.predict(X_test)
print(classification_report(y_test, pred_svc2))
```
