```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import matplotlib.font_manager as fm

font_location = 'C:/Windows/Fonts/H2GTRM.TTF' 
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font', family = font_name)
```

```python
movie = pd.read_excel('movie.xlsx')
movie['개봉일'] = pd.to_datetime(movie['개봉일'])
movie['개봉월'] = pd.DatetimeIndex(movie['개봉일']).month
```

```python
movie
```

```python
genre = movie['영화명'].groupby(movie['장르']).count().sort_values()
genre = genre.dropna()
```

```python
genre.plot.barh(color=sns.color_palette('summer_r',21))
plt.title('장르별 영화 수')
```

```python
nation = movie['영화명'].groupby(movie['국적']).count().sort_values()
nation[-5:].plot.barh(color=sns.color_palette('summer_r',21))
plt.title('국가별 영화 수')
```

```python
level = movie['영화명'].groupby(movie['개봉월']).count().sort_values()
level.plot.barh(color=sns.color_palette('summer_r',21))
plt.title('월별 영화 수')
```

```python
dist = movie['영화명'].groupby(movie['배급사']).count().sort_values()
dist[-10:].plot.barh(color=sns.color_palette('summer_r',21))
plt.title('배급사별 영화 수')
```

```python
director = pd.DataFrame(movie)
director = movie['영화명'][movie['전국관객수']>3000000].groupby(movie['감독']).count().sort_values()
director = pd.Series(director)
director[-10:].plot.barh(color=sns.color_palette('summer_r',21), xticks=[0,1,2,3,4])
plt.title('어떤 감독이 몇편이나 흥행 시켰을까?')
```

```python
director_num = pd.DataFrame(movie)
director_num = director_num['전국관객수'].groupby(director_num['감독']).sum().sort_values()
director_num[-10:].plot.barh(color=sns.color_palette('summer_r',21), xlim=(15000000,35000000))
plt.title('감독별 총 관객수')
```

```python
director_mean = pd.DataFrame(movie)
director_mean = director_mean['전국관객수'].groupby(director_mean['감독']).mean().sort_values()
director_mean[-10:].plot.barh(color=sns.color_palette('summer_r',21), xlim=(6000000,15000000))
plt.title('감독별 평균 관객수')
```

```python
genre_sum = pd.DataFrame(movie)
genre_sum = genre_sum['전국관객수'].groupby(genre_sum['장르']).sum().sort_values()
genre_sum.plot.barh(color=sns.color_palette('summer_r',21))
plt.title('장르별 관객수')
```

```python
genre_mean = pd.DataFrame(movie)
genre_mean = genre_mean['전국관객수'].groupby(genre_mean['장르']).mean().sort_values()
genre_mean.plot.barh(color=sns.color_palette('summer_r',21))
plt.title('장르별 관객수')
```

```python
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
```

```python
str_list = []
for col_name, col_value in movie.iteritems():
    if type(col_value[1]) == str:
        str_list.append(col_name)
```

```python
num_list = movie.columns.difference(str_list)
movie_num = movie[num_list]
movie_num.drop(movie_num.columns[1],axis=1,inplace=True)
movie_num.head()
```

```python
movie_num.drop(movie_num.columns[4],axis=1,inplace=True)
```

```python
X = movie_num.values
x = StandardScaler().fit_transform(X)
normalized_movie = pd.DataFrame(x)
normalized_movie = normalized_movie.rename(columns={0:'month',1:'quater',2:'Seoul viewer',3:'Seoul sales',4:'movie price',5:'viewers', 6:'sales',7:'screens'})
```

```python
corr = normalized_movie.corr()
```
