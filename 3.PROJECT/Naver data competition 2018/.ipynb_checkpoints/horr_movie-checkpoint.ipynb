# 영화 데이터 분석

```python
import pandas as pd
```

```python
movie = pd.read_excel('movie_data.xlsx')
```

```python
movie
```

```python
cols = [0, 4]
movie.drop(movie.columns[cols],axis=1,inplace=True)
```

```python
movie.dropna(inplace=True)
```

```python
movie_horr = movie.loc[movie['장르']=='공포(호러)'].dropna()
```

```python
movie_thriller = movie.loc[movie['장르']=='스릴러'].dropna()
```

```python
movie_summer = pd.concat([movie_horr, movie_thriller])
```

```python
print('1분기 공포&스릴러 영화수: {0}'.format(movie_summer['영화명'][movie_summer['분기']==1].count()))
print('2분기 공포&스릴러 영화수: {0}'.format(movie_summer['영화명'][movie_summer['분기']==2].count()))
print('3분기 공포&스릴러 영화수: {0}'.format(movie_summer['영화명'][movie_summer['분기']==3].count()))
print('4분기 공포&스릴러 영화수: {0}'.format(movie_summer['영화명'][movie_summer['분기']==4].count()))
```

```python
print('1분기 공포&스릴러 영화 관객수: {0}, 퍼센트: {1:.2f}%'.format(movie_summer['전국관객수'][movie_summer['분기']==1].sum(),movie_summer['전국관객수'][movie_summer['분기']==1].sum() / movie_summer['전국관객수'].sum() * 100))
print('2분기 공포&스릴러 영화 관객수: {0}, 퍼센트: {1:.2f}%'.format(movie_summer['전국관객수'][movie_summer['분기']==2].sum(),movie_summer['전국관객수'][movie_summer['분기']==2].sum() / movie_summer['전국관객수'].sum() * 100))
print('3분기 공포&스릴러 영화 관객수: {0}, 퍼센트: {1:.2f}%'.format(movie_summer['전국관객수'][movie_summer['분기']==3].sum(),movie_summer['전국관객수'][movie_summer['분기']==3].sum() / movie_summer['전국관객수'].sum() * 100))
print('4분기 공포&스릴러 영화 관객수: {0}, 퍼센트: {1:.2f}%'.format(movie_summer['전국관객수'][movie_summer['분기']==4].sum(),movie_summer['전국관객수'][movie_summer['분기']==4].sum() / movie_summer['전국관객수'].sum() * 100))
```

```python
dummies = pd.get_dummies(movie_summer['분기'], prefix='분기')
dummies.head()
```

```python
movie_result = pd.concat([movie_summer, dummies], axis=1)
```

```python
movie_result.corr()
```

```python
movie_summer.reset_index(inplace=True)
movie_summer.index = movie_summer.index + 1
```
