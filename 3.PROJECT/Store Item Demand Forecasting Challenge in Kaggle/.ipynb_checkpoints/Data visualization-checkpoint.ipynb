## Add day of week
### Monday=0, Tuesday=1, Wednesday=2, Thursday=3, Friday=4, Saturday=5, Sunday=6

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['day'] = data['date'].dt.dayofweek
data.head()
```
