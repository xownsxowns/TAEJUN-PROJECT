```python
# 요일별 수익률 비교
import pandas as pd
from antQuant.utils import DataManager
```

```python
celltrion = DataManager()
celltrion_stock = celltrion.get_daily_ohlcv('068270')
```

```python
weekday = celltrion_stock.index.weekday
```

```python
celltrion_stock['weekday'] = weekday
```

```python
celltrion_stock
```

```python
# 전날 종가가 상승했을 때 종가와 다음날 사이 수익률 비교
profit = 0
number_of_profit = 0
for i in range(len(celltrion_stock.index)):
    if celltrion_stock.close[i-1] > celltrion_stock.open[i-1]:
        pf = celltrion_stock.open[i] - celltrion_stock.close[i-1]
        profit = profit + pf
        number_of_profit += 1
    else:
        continue
```

```python
# 수익이 일어날 확률
percent_profit = (number_of_profit / len(celltrion_stock.index)) * 100
print("수익이 일어날 확률은 {0}".format(percent_profit))
```

```python
# 전날 종가가 상승했을 때 종가와 다음날 사이 수익률 비교
profit = 0
number_of_profit = 0
for i in range(len(celltrion_stock.index)):
    if celltrion_stock.close[i-1] < celltrion_stock.open[i-1]:
        pf = celltrion_stock.open[i] - celltrion_stock.close[i-1]
        profit = profit + pf
        number_of_profit += 1
    else:
        continue
```

```python
# 수익이 일어날 확률
percent_profit = (number_of_profit / len(celltrion_stock.index)) * 100
print("수익이 일어날 확률은 {0}".format(percent_profit))
```

```python
close_open = []
for i in range(len(celltrion_stock.index)):
    p = celltrion_stock.close[i] - celltrion_stock.open[i]
    close_open.append(p)
```

```python
celltrion_stock['close_open']=close_open
```

```python
celltrion_stock
```

```python
mon_profit = 0
tue_profit = 0
wen_profit = 0
thu_profit = 0
fri_profit = 0

for i in range(len(celltrion_stock.index)):
    if celltrion_stock.weekday[i] == 0:
        mon_profit = mon_profit + celltrion_stock.close_open[i]
    elif celltrion_stock.weekday[i] == 1:
        tue_profit = mon_profit + celltrion_stock.close_open[i]
    elif celltrion_stock.weekday[i] == 2:
        wen_profit = wen_profit + celltrion_stock.close_open[i]
    elif celltrion_stock.weekday[i] == 3:
        thu_profit = thu_profit + celltrion_stock.close_open[i]
    elif celltrion_stock.weekday[i] == 4:
        fri_profit = fri_profit + celltrion_stock.close_open[i]
```

```python
print(mon_profit, tue_profit, wen_profit, thu_profit, fri_profit)
```
