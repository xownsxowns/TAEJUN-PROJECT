```python
# 전날 종가가 고가 또는 저가와 가까울 때의 다음날 시가 사이의 수익률 비교
import pandas as pd
from antQuant.utils import DataManager
```

```python
stock = DataManager()
celltrion_stock = stock.get_daily_ohlcv('068270')
celltrion_stock.head()
```

```python
# 전날 종가가 고가와 가까울 때의 다음날 시가 사이의 수익률 비교
import numpy as np
returns = []
for i in range(len(celltrion_stock.index)):
    if i == len(celltrion_stock.index)-1:
        p = None
        returns.append(p)
    else:
        p = celltrion_stock.open[i+1]-celltrion_stock.close[i]
        returns.append(p)

celltrion_stock['returns'] = returns
celltrion_stock['returns'] = np.log(celltrion_stock['returns'])
celltrion_stock = celltrion_stock[(celltrion_stock['returns']!=np.inf)&(celltrion_stock['returns']!=-np.inf)]
celltrion_stock.dropna()
```
