```python
# 전날 종가와 다음날 시가 사이 수익률 비교 - 기초통계(평균, 중간값, 표준편차), 확률, 손익비
# 전날 종가에 사서 다음날 시가에 판다면?
import pandas as pd
from antQuant.utils import DataManager
```

```python
skhynix = DataManager()
skhynix_stock = skhynix.get_daily_ohlcv('000660')
```

```python
skhynix_stock
```

```python
close_open = []
for i in range(len(skhynix_stock.index)):
    pf = skhynix_stock.open[i] - skhynix_stock.close[i-1]
    close_open.append(pf)
```

```python
skhynix_stock['close_open']=close_open
```

```python
skhynix_stock = skhynix_stock[1:]
```

```python
skhynix_stock
```

```python
# 통계값 구하기: numpy 이용
import numpy as np
mean_close_open = np.mean(skhynix_stock.close_open)
median_close_open = np.median(skhynix_stock.close_open)
std_close_open = np.std(skhynix_stock.close_open)
```

```python
print("mean of subtracting from close to open is {0}".format(mean_close_open))
print("median of subtracting from close to open is {0}".format(median_close_open))
print("std of subtracting from close to open is {0}".format(std_close_open))
```

```python
# 수익이 날 확률!
count = 0
for i in range(len(skhynix_stock.index)):
    if skhynix_stock.close_open[i] > 0:
        count += 1

percentile_count = (count / len(skhynix_stock.index)) * 100
print("Percent of the number of profit is {0}".format(percentile_count))
```

```python
# 손익 계산
profit = 0
loss = 0
for i in range(len(skhynix_stock.index)):
    if skhynix_stock.close_open[i] > 0:
        profit = profit + skhynix_stock.close_open[i]
    elif skhynix_stock.close_open[i] < 0:
        loss = loss + skhynix_stock.close_open[i]
        
print("profit is {0}".format(profit))
print("loss is {0}".format(loss))
```
